#Introduction

Diabetes has become a widespread issue in the United States. The incidence of diabetes has risen from "180 million in 1980," to "422 million in 2014" (Roglic, 2016). In adults the incidence has almost doubled "from 4.7% in 1980 to 8.5% in 2014" and is the highest in countries with high income (Roglic, 2016). Diabetes can be characterized by the prevalence of high blood glucose levels, which, due to the lack of insulin produced in the pancreas, glucose remains unmetabolized within the cells (Roglic, 2016). In addition, cell ineffectively use the insulin produced, resulting in high blood glucose levels. There are three classifications of diabetes: Type 1, Type 2, and gestational diabetes. In Type 1 diabetes, there is no insulin being produced in the pancreas (Roglic, 2016). In Type 2, insulin production decreases due to the resistance of cells against "the action of insulin that is being produced" in the body (Roglic, 2016). Gestational diabetes occurs during pregnancy, this type can cause complications with pregnancy as well as "increase the risk of type 2 diabetes" (Roglic, 2016).

There are many risk factors in the development of diabetes, in Type 2 the risk factors are clearer and in Type 1. The former includes genetics, "age, overweight and obesity  and physical inactivity, smoking," however the largest factor that increases risk is "body fat" (Roglic, 2016). When reviewing the effects of BMI on the chance of diabetes at 18 years old the risk of diabetes jumps from “12.2% to 74.4%” and lifetime risk increased from “8.7% to 26.7%” for overweight and obese women (Narayan, et al., 2007). When investigating another factor, it can be found that “a worldwide increase in the prevalence of type 2 diabetes is now being realized in the pregnancy context with apparently similar or even worse outcomes to type 1 diabetes” (McCance, 2011). As related to age, the incidence of diabetes increase with age, especially those who were women and single between the ages of 35 to 64 years old (Choi, Shi, 2001). BMI and number of pregnancies are some of the medical predictors that the model uses to predict diabetes. Many different medical analyses have determined that BMI and number of pregnancies, among many other factors including insulin level and age, have a pronounced effect on the chances of diabetes in a patient.

When assessing the prevalence of diabetes in communities of Pima Indians, it was found that there existed an "extraordinary high prevalence of hyperglycemia and the specific complications of diabetes... in abundance in the Pima Indians" (Bennett, et al., 1976). In fact, the Pima Indians that reside in Arizona recorded the highest "prevalence and incidence of non-insulin dependent diabetes of any geographically-defined population" (Knowler, et al., 1981).

Since the Pima Indians have such an alarming prevalence of diabetes, it remains a relevant question as to how best to predict diabetes in this population. This can be done with specific diagnostic markers, thus analyzing these markers to see which prove to be the most relevant or reliable in these predictions can help to develop interventions for this disease.

Logistic Regression fits a linear equation to our data and uses a logistic function to predict probabilities for binary classification based on our 8 features. Our remaining three classical machine learning models are based on Decision Tree, which splits the dataset into subsets based on feature thresholds, aiming to maximize class purity at each node. Gradient Boosting uses an ensemble of decision trees, aiming to correct the errors of the previous generation by minimizing the loss function. Random Forest uses multiple decision trees, each trained on a random subset of the total data, and aggregates the predictions in order to minimize loss, prevent overfitting and increase accuracy.

##Performance Evaluation Metrics


Once we create and train these machine learning models on our dataset, we need to know how to assess them and their performance. We will use seven metrics during this study. The most basic of which is accuracy, which simply measures how often a model is correct at predicting the diabetes status of a person. However, accuracy is not a resistant measurement, and does not give any information about what cases a model is getting correct or incorrect. We use precision, log loss, ROC-AUC, recall (sometimes referred to as sensitivity) and specificity to get more information about what the model is getting correct and incorrect. Finally, we compute the balanced accuracy, a more resistant method of comparing the different models against each other. With these metrics, we can more get a better picture of what each models excels and struggles with, and which are performing the best overall.

In the equations below, TP represents the number of true postives, ie the number of people the model predicts have diabetes that actually have diabetes. TN, or true negative count would then be the number of people the model predicts do not have diabetes, who do in fact not have diabetes. Conversely, FP or false positives are individuals who the model predicts have diabetes that do not. Finally, FN or false negatives occur where the model guesses a person does not have diabetes, when they do. The following equations then govern these metrics:

$$\text{Accuracy} = \frac{TP+TN}{TP + FP + TN + FN}$$

$$\text{Precision} = \frac{TP}{TP + FP}$$

$$\text{Recall} = \frac{TP}{TP + FN}$$

$$\text{Specificity} = \frac{TN}{TN + FP}$$

$$\text{Balanced Accuracy} = \frac{\frac{TP}{TP + FN} + \frac{TN}{TN + FP}}{2}$$




$$\text{Log Loss} = - \frac{1}{n} \sum_{i=1}^{n} \left[ y_i \log(p_i) + (1 - y_i) \log(1 - p_i) \right]
$$

In the equation for log loss, n is the number of samples, yi is the class label, Pi is the probability that the sample fits one class, for the i-th sample. The 1/n out front means that this equation normalizes the loss by the number of samples. An ideal log loss would be 0 in this case, meaning that the model was correct every single time.

#Experiment Setups and Result Discussion

##Parameters and Experiment Settings
**Experimental Setting:** For our experiment, we used Google Colab on Microsoft Windows 11. We used Python 3.10, utilizing various tools and packages like sklearn and keras tensorflow (Pedregosa et al., 2011), (Abadi et. al. 2016).

In order to assess the accuracy of our models, we need to utilize some system that would allow us to know the target value with full confidence (ie whether or not a particular person has diabetes). In order to do this, we utilized the train-test-split method. We used the sklearn library to do so (Pedregosa et al., 2011). This method separates the data into two categories, training and testing (or validation as it may be referred to later) data. 80% of the data is used for training, and the remaining 20% is used as validation. The data is organized into these groups randomly. Our models used the validation data to test their accuracy, since the model knows the target boolean value of that data.


**Model Instantiation and Parameters:**
For our six models, five used Sklearn’s models. Those five were Logistic Regression, SVM, Random Forest, Gradient Boosting and Decision Tree. The Neural Network model used keras and tensorflow. All five of the former models used the same random state, for consistency, of random_state =10.


Logistic Regression used the ‘Limited Memory Broyden Fletcher Goldgarb Shanno Algorithm’, or lmbfg. It is a quasi-Newton method that aims to approximate the inverse of the Hessian matrix, and minimizes log loss. We leave the parameter C (inverse regulation strength) at 1 to balance low training error and low regularization penalty. We used a tolerance of 0.0001 for Logistic Regression.
For the Singular Value Decomposition model, we similarly use an inverse regulation strength of 1, and a tolerance of 0.001. Here, SVM uses a slightly higher tolerance due to the difference in complexity of data fitting. We use the ‘Radial Basis Function Kernal’, and allow the gamma to adjust to it. Using the formula G = $\frac{1}{n_{features}\cdot X.var()}$.
Since Random Forest is a collection of Decision Trees, here, both Random Forest and Decision Tree here use the criterion of entropy. This parameter aims to maximize information gain after each split. Here, entropy, H = $- \sum_{i=1}^{n} p_i \log_2(p_i)$, for $p_i$ equal to the proportion of samples belonging to class
i in the node. The Information Gain after each split can then be quantified as $G = H_{parent}-\sum_{Children}\frac{S_c}{S_p} \cdot H_{Child}$, where $S_c$ is the size of the child and $S_p$ is the size of the parent. In Random Forest, each of the decision trees within the model use the entropy criterion.
The last sklearn mode, Gradient Boosting, uses the criterion of Friedman_MSE, an optimized version of MSE, mean squared error. It is faster and has better splits than just using MSE. We also use a learning rate of 0.1, and aim to minimize the log loss.
The Neural Network model functions slightly differently. We use Keras and Tensorflow, with 50 epochs, the ADAM optimizer, a learning rate of 0.001, and a loss function of Binary Cross Entropy. As far as the structure of the model is concerned, the input layer is a fully connected layer with 64 perceptrons using the ReLU activation function, followed by another fully connected layer with 32 perceptrons also using the ReLU activation function. The output layer consists of 1 perceptron with a sigmoid activation function.

Recall measures the ability of the model to correctly predict true positive values. It is similar to precision, but not exactly the same. Precision measures how many of the model’s predicted positive values were actually positive, and recall measures how many of the total positive values the model was able to identify. Interestingly, Decision Tree did not do the worst on this aspect, but SVM, and Random Forest did the best.


The final metric measured for our models was balanced accuracy, which is the average of specificity and recall. Balanced accuracy can be a more objective measurement for accuracy in the case of imbalance datasets. Our datasets has more people without diabetes (negatives) than people with diabetes (positive values), and balanced accuracy ensures that our model does well at predicting both classes equally, and not just the majority class. Our scores for balanced accuracy are lower than our scores for accuracy, which is what we expect given the values of the other metrics as described above.
