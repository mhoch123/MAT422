#2.3.1 Joint probability distributions

**Two Discrete Random Variables:** If $X$ and $Y$ are both discrete random variables in sample space $S$ then the joint probability mass function $p(x,y)$ is defined for each pair of numbers $(x,y)$ by $p(x,y)=P(X=x,Y=y)$

**Marginal Probability Mass Function:** for X, denoted by $p_{X}(x)$, is given by $p_{X}(x)=\sum\limits_{y:p(x,y)>0}p(x,y)$ for each possible value of x. Similarly for Y, $p_{Y}(y)=\sum\limits_{x:p(x,y)>0}p(x,y)$ for each value of y.

**Two Continuous Random Variables:** If $X$ and $Y$ are continuous random variables, then for any two-dimensional set A, $P[(X,Y)\in A]=P(a\geq X\geq b, c\geq Y\geq d)= \int_{a}^{b}\int_{c}^{d}f(x,y)dydx$

**Independent Random Variables:** Two random variables $X$ and $Y$ are said to be independent if for every pair of x and y values, $p(x,y)=p_{X}(x)\cdot p_{Y}(y)$ when $X$ and $Y$ are discrete or $f(x,y)=f_{X}(x)\cdot f_{Y}(y)$ when $X$ and $Y$ are continuous.

```python
import numpy as np
import matplotlib.pyplot as plt
import random

np.random.seed(2)
N =500
x = np.random.randn(N)
y = np.random.randn(N)**2
plt.scatter(x,y)
```

Here we see a joint probability distribution.

#2.3.2. Correlation and dependence

**Covariance:** The covariance between two random variable's $X$ and $Y$ is $Cov(X,Y)=E[(X,\mu_{x})(Y-\mu_{y})] =
\left\{ \begin{array}{rcl}
\sum_{x} \sum_{y}(x-\mu_{X})(y-\mu_{Y})p(x,y) & \mbox{X,Y discrete} \\ \int_{-\infty}^{\infty} \int_{-\infty}^{\infty}(x-\mu_{X})(y-\mu_{Y})f(x,y)dxdy & \mbox{X,Y continuous}
\end{array}\right.$

**Correlation Coefficient:**Is denoted by $Corr=(X,Y),\rho_{X,Y}$ or $\rho$ is defined as $\rho_{X,Y} =  \frac{Cov(X,Y)}{\sigma_{X}⋅\sigma_{Y}}$



```python
x = np.random.rand(1,100)
y = x
y2 = np.random.rand(1,100)
x2 = np.linspace(-10,10,100)
y3 = x2**2

fig, ax = plt.subplots(1,3, figsize = (12,3))
ax[0].plot(x,y,'bo')
ax[1].plot(x,y2,'bo')
ax[2].plot(x2,y3,'bo')
plt.show

r1 = np.corrcoef(x,y)
r2 = np.corrcoef(x,y2)
r3 = np.corrcoef(x,y3)
print('Correlation 1:', r1[0,1])
print('Correlation 2:', r2[0,1])
print('Correlation 3:', r3[0,1])
```

Here are three examples of different graphs with different correlations. The first one has a completely positive correlation. The next, random data, has almost zero correlation. Finally the third, also has a low correlation, even though we can see that there is a clear relationship between the x and the y variable, showing that correlation isn't always the best metric to use.

#2.3.3. Random samples

**Random Samples:** Let $X_1,X_2,X_3,\ldots,X_n$ be a random sample from a distribution with mean value $\mu$ and a standard deviation of $\sigma$. Then 1. $E(X)=\mu_{x} =\mu$ 2.$V(X)= \sigma_{X}^{2} = \frac{\sigma^{2}}{n}$ and $\sigma_X = \frac{\sigma}{\sqrt{n}}$

**Central Limit Theorem:** (CLT) indicates that the properly normalized sum of independent random variables tends toward a normal distribution even if the original variables themselves are not normally distributed

```python
population = np.arange(500)
population = tuple(population)
sample = random.sample(population,50)
sample = sorted(sample)
data = np.random.normal(75,10,500)
i = 0
sample_data = []
while i < 50:
  sample_data.append(data[sample[i]])
  i = i + 1
plt.hist(sample_data)


```

Here we see a random sample of 10% of a population. The population is normally distributed, so we expect the same to also be roughly normal, with a mean of 75 and a standard deviation of 10. Our sample is smaller than the population, so the data may look slighly less normal because of the smaller sample size.
